```python
numerical_corelation = df1.select_dtypes(exclude='object').corr()
plt.figure(figsize=(20,20))
plt.title("High Corelation")
sns.heatmap(numerical_corelation>0.7, annot=True, square=True)
```

```python
numerical_data = df1.copy()
```

```python
print(numerical_data)
```

```python
print(numerical_data.columns)
```

```python
fig = plt.figure(figsize=(12,18))
for i in range(len(numerical_data.columns)):
    fig.add_subplot(9,4,i+1)
    sns.distplot(numerical_data.iloc[:,i].dropna(), rug=True, hist=True, label='UW', kde_kws={'bw':0.1})
    plt.xlabel(numerical_data.columns[i])
plt.tight_layout()
plt.show()
```

```python
fig3 = plt.figure(figsize=(12,18))
for i in range(len(numerical_data.columns)):
    fig3.add_subplot(9, 4, i+1)
    sns.scatterplot(numerical_data.iloc[:, i],df1['class'])
plt.tight_layout()
plt.show()
```
